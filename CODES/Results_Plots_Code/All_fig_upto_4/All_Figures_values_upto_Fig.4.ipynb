```python
def return_df(df, file, val):
    ls = ['lr', 'rr', 'knn','gpr','dt','rf',
          'etr','dnn']
    new_df = pd.DataFrame(columns = ls) 
    for i in range(len(ls)):
        arr = [float(s) for s in df[val][i][1:-1].split(',')]
        new_df[ls[i]] = arr
    return new_df
```

```python
def return_df_mape(df, file, val):
    ls = ['lr', 'rr', 'knn','gpr','dt','rf',
          'etr','dnn']
    new_df = pd.DataFrame(columns = ls) 
    for i in range(len(ls)):
        arr = [float(s) for s in df[val][i][1:-1].split(',')]
        new_df[ls[i]] = arr
    return new_df
# return_df_mape(temp_df, file = 'none', val = 'mape')
```

```python
def find_y_r2(df):
    y = []
    for i in range(len(df.columns)):
        y_temp = df[df.columns[i]].tolist()
        y.append(np.mean(y_temp))
    # print(y)    
    return y
```

```python
def find_y_mape(df):
    y = []
    for i in range(len(df.columns)):
        y_temp = df[df.columns[i]].tolist()
        y_temp = [x * 100 for x in y_temp]
        y.append(np.median(y_temp))
    # print(y)    
    return y
```

```python
def find_y_mse(df):
    y = []
    for i in range(len(df.columns)):
        y_temp = df[df.columns[i]].tolist()
        y.append(np.mean(y_temp))
    return y
```
